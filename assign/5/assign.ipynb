Ένα  ολοκληρωμένο κύκλωμα έχει τετραγωνικό σχήμα με πλευρά $L$ και πάχος $w$. Κατά την λειτουργία παράγεται θερμότητα $\dot{q}_S$ ομοιόμορφα στο εσωτερικό του. Υπολογίστε την θερμοκρασία κατά μήκος του πάχους, αν και οι δύο πλευρές είναι εκτεθειμένες σε ροή αέρα, ενώ οι παράπλευρες επιφάνειες μπορούν να θεωρηθούν μονωμένες. Ο υπολογισμός αυτός είναι χρήσιμος για τον έλεγχο της μέγιστης θερμοκρασίας, η οποία δεν πρέπει να ξεπερνά τους $85°C$ σε
εμπορικές εφαρμογές.

Το πρόβλημα αυτό είναι μονοδιάστατο μόνιμης κατάστασης και περιγράφεται από την διαφορική εξίσωση: 

$$
k\frac{d^2T}{dx^2}+S=0
$$

όπου $S$ η πηγή θερμότητας στην μονάδα του όγκου:

$$
S=\frac{\dot{q}_S}{V}
$$

Για την επίλυσή του, εφαρμόστε πεπερασμένες διαφορές χωρίζοντας το πάχος σε $n=10$ διαστήματα σταθερού βήματος $Δx=\dfrac{w}{n}$.

$$
-k\frac{T_1-T_0}{Δx}=h(T_{a}-T_0)+S\frac{Δx}2
$$

$$
-k\frac{T_n-T_{n-1}}{Δx}=h(T_n-T_{a})-S\frac{Δx}2
$$
